```python
#importing the necessary libraries and methods
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains

import time
```

```python
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
# My chromedriver is outdated so installed chromedriver manager
```

```python
driver.maximize_window() # maximizing the window

driver.get('https://www.cars24.com/')
time.sleep(1)
```

```python
#Clicking on view all cars
view_all_cars = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div[2]/div/div[2]/button[1]')
view_all_cars.click()
time.sleep(2)
ignore_setup = driver.find_element(By.XPATH,'/html/body/div[1]/div[8]/div/div/h3/div') # clicking on the setup button
ignore_setup.click()
time.sleep(1)
```

```python
# scraping the links of each car
car_urls = []
car_link_tags = driver.find_elements(By.XPATH,'//a[@class="_9Ue0B"]')
for i in range(0,1):
    for l in car_link_tags:
        time.sleep(1)
        car_urls.append(l.get_attribute('href'))
        time.sleep(1)
    
```

```python
print(len(car_urls))
```

```python
# Creating Instances

brand = []
model = []
variant = []
location = []
accident = []
transmission = []
owners = []
fuel_type = []
kilometers = []
price = []
length = []
airbags = []
year = []
```

```python
# Scraping the prices
price_tags = driver.find_elements(By.XPATH,'//div[@class="_7udZZ"]')
for i in price_tags:
    price.append(i.text.split('₹')[1])
```

```python
for i in car_urls:
    driver.get(i)
    time.sleep(4)
    
    # brand
    try:
        for a in driver.find_elements(By.XPATH,'//h2[@class="_2geSF"]'):
            brand.append(a.text.strip().split()[1])
    except NoSuchElementException:
        brand.append('-')
        
    # year
    try:
        for y in driver.find_elements(By.XPATH,'//h2[@class="_2geSF"]'):
            year.append(y.text.strip().split()[0])
    except NoSuchElementException:
        year.append('-')
        
    # model
    try:
        for i in driver.find_elements(By.XPATH,'//h2[@class="_2geSF"]'):
            model.append(i.text.strip().split()[2])
    except NoSuchElementException:
        model.append('-')
        
    # variant
    try:
        for v in driver.find_elements(By.XPATH,'//h2[@class="_2geSF"]'):
            variant.append(i.text.strip().split()[3])
    except NoSuchElementException:
        variant.append('-')
    
    # location
    try:
        for b in driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/div[3]/section/div/div/div/div[1]/p[2]/span'):
            location.append(b.text)
    except NoSuchElementException:
        location.append('-')
    
    # kilometers
    try:
        for c in driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/section[1]/div/div/div/div[1]/div/ul/li[3]/strong'):
            kilometers.append(c.text.strip().split()[0])
    except NoSuchElementException:
        kilometers.append('-')
    
        
    # accident    
    try:
        for e in driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/section[1]/div/div/div/div[1]/div/ul/li[1]/strong'):
            accident.append(e.text)
    except NoSuchElementException:
        accident.append('-')
    
    # transmission
    try:
        for f in driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/section[1]/div/div/div/div[1]/div/ul/li[6]/strong'):
            transmission.append(f.text)
    except NoSuchElementException:
        transmission.append('-')
    
    # owners
    try:
        for g in driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/section[1]/div/div/div/div[1]/div/ul/li[2]/strong'):
            owners.append(g.text)
    except NoSuchElementException:
        owners.append('-')
    
    # fuel_type
    try:
        for h in driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/section[1]/div/div/div/div[1]/div/ul/li[4]/strong'):
            fuel_type.append(h.text)
    except NoSuchElementException:
        fuel_type.append('-')
    
    # scrolling the page
    for _ in range(8):
        driver.execute_script("window.scrollBy(0,8)")
    
    # clicking on features
    # scraping airbags data
    click_features = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/section[1]/div/div/div/div[1]/ul/li[2]').click()
    try:
        for k in driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/section[1]/div/div/div/div[1]/div/ul/li[1]/strong'):
            airbags.append(k.text)
    except NoSuchElement:
        airbags.append('-')
    
    time.sleep(3)
    
    # clicking on view features
    # scraping length data
    view_features = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/section[1]/div/div/div/div[2]/div/button').click()
    try:
        for j in driver.find_elements(By.XPATH,'/html/body/div[5]/div/div/div/div[3]/ul/li[1]/ul/li[7]/strong'):
            length.append(j.text)
    except NoSuchElementException:
        length.append('-')
               
```
